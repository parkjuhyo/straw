```python
import os
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
import shutil

#폴더 경로 지정
filepath = 'Z:/촬영데이터/0810수확_0815촬영/'
#폴더 속 파일 리스트로 불러오기
folder = os.listdir('Z:/촬영데이터/0810수확_0815촬영/')

# 이동할 폴더 경로
new_filepath ='Z:/촬영데이터/0810수확_0815촬영/'

# 폴더 속 파일 갯수만큼 반복
for i in range():
    # 이미지 불러오기 ( filepath 폴더 경로 + folder[i+1] 파일명)
    im = Image.open(filepath + folder[i])
    # 불러온 이미지를 배열로 만들기
    im = np.array(im)
    # 불러온 이미지의 픽셀값이 160 이상인 것은 0으로 만들기(0=검정색)
    #im[im <= 160] = 0
    # 이미지 배열을 1차원으로 바꾸기
    im = np.ravel(im)
    # 전체 픽셀값을 더한 것 / 픽셀갯수 
    im = np.mean(im)

    print(folder[i], ":", im)
    

    if im <= 144:
        shutil.move(filepath + folder[i], new_filepath + folder[i])
```

```python
from PIL import Image

def image_crop( image_path, filename, save_path ):
    
    #원본 이미지 불러오기
    img = Image.open(image_path + filename)
    #원본 이미지 사이즈(높이,넓이)
    (img_h, img_w) = img.size
    print("원본 이미지 사이즈 : ",img.size)
    
    #crop할 사이즈 입력
    grid_w = 750 #넓이
    grid_h = 1000 #높이
    range_w = (int)(img_w/grid_w)
    range_h = (int)(img_h/grid_h)
    print(range_w, range_h)
    
    i = 0
    
    for w in range(range_w):
        for h in range(range_h):
            bbox=(h*grid_h,w*grid_w,(h+1)*(grid_h),(w+1)*(grid_w))
            print(h*grid_h,w*grid_w,(h+1)*(grid_h),(w+1)*(grid_w))
            
            crop_img = img.crop(bbox)
            
            #파일이름지정
            filename = filename.split(".")[0]
            fname = "{}.jpg".format("%s_%d"%(filename,i+1))
            
            #저장경로 + 파일이름
            savename = save_path + fname
            
            #이미지 저장
            crop_img.save(savename)
            print('save_file'+ savename +'....')
            
            i += 1
```

```python
image_path = 'Z:/촬영데이터/0810수확_0815촬영/'  #이미지경로
filename = os.listdir(image_path) #파일명
filename.sort()

save_path = 'Z:/정제데이터/0810수확_0815촬영/' #저장경로

# for i in range(len(filename)):
#   image_crop(filename[i] , save_path)

for i in range(len(filename)):
  image_crop(image_path , filename[i] , save_path )
```
